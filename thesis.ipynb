```python
import tifffile as tiff
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMG_DIM = 128
```

```python
index = np.random.randint(0, 500)

img = tiff.imread(f"./dataset/train_data_{IMG_DIM}/{index}_0.tif").astype(np.float32)
img = 255 * (img - img.min()) / (img.max() - img.min())
if img.shape[-1] == 4:
    img = img[:, :, :3]


mask = tiff.imread(f"./dataset/label_data_{IMG_DIM}/{index}_mask.tif").astype(
    np.float32
)

# Plot image and mask side by side
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(img.astype(np.uint8))
plt.title("Image")
plt.subplot(1, 2, 2)
plt.imshow(mask)
plt.title("Mask")
plt.show()
```
